```python
from sklearn.preprocessing import MinMaxScaler
values = data[['종가']].values.reshape(-1,1)
values = values.astype('float32')
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)
```

```python
train_size = int(len(scaled) * 0.8)
val_size = len(scaled) - train_size

test_size = len(scaled) - train_size
train, test = scaled[0:train_size,:], scaled[train_size:len(scaled),:]
print(len(train), len(test))
```

```python
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i:(i + look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    print(len(dataY))
    return np.array(dataX), np.array(dataY)
```

```python
look_back = 7
trainX, trainY = create_dataset(train, look_back)
#valX, valY = create_dataset(val, look_back)
testX, testY = create_dataset(test, look_back)
```

```python
print(trainX.shape)
print(trainY.shape)
#print(valX.shape)
#print(valY.shape)
print(testX.shape)
print(testY.shape)
```

```python
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
#valX = np.reshape(valX, (valX.shape[0], 1, valX.shape[1]))
testX = np.reshape(testX, (testX.shape[0],testX.shape[1], 1 ))
```

```python
trainX.shape
```

### LSTM

```python
from matplotlib import pyplot
```

```python
model = Sequential()
model.add(Bidirectional(LSTM(100, input_shape=(7, trainX.shape[2]))))

model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')
history = model.fit(trainX, trainY, epochs=100, batch_size=100, validation_data=(testX, testY), verbose=0, shuffle=False)
```

```python
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```

```python
yhat = model.predict(testX)

plt.figure(figsize=(20,10))

pyplot.plot(yhat, label='predict')
pyplot.plot(testY, label='true')
pyplot.legend()
pyplot.show()
```

```python
yhat_inverse = scaler.inverse_transform(yhat.reshape(-1,1))
testY_inverse = scaler.inverse_transform(testY.reshape(-1, 1))
```

```python
from math import *
rmse = sqrt(mean_squared_error(testY_inverse, yhat_inverse))
print('Test RMSE: %.3f' % rmse)
```

## stacked

```python
multi_model = Sequential()
multi_model.add(LSTM(100, input_shape=(trainX.shape[1], trainX.shape[2]),return_sequences=True))

multi_model.add(LSTM(64))
multi_model.add(Dense(1))
multi_model.compile(loss='mae', optimizer='adam')
multi_history = multi_model.fit(trainX, trainY, epochs=300, batch_size=100, validation_data=(testX, testY), verbose=0, shuffle=False)
```

```python
pyplot.plot(multi_history.history['loss'], label='multi_train')
pyplot.plot(multi_history.history['val_loss'], label='multi_test')
pyplot.legend()
pyplot.show()
```

```python
yhat = multi_model.predict(testX)
pyplot.plot(yhat, label='predict')
pyplot.plot(testY, label='true')
pyplot.legend()
pyplot.show()
```
